## Seleccion de Caracteristicas 

```python
from sklearn.feature_selection import SelectKBest
import numpy as np
```

```python
x = df[['ID_CLIENTE','EDAD','MESESINACTIVIDAD','CANTADICIONALES','LIMITE_DISP_TC',
        'ANTIGUEDAD','INACTIVIDADCOMPRA','GRUPO_AFINIDAD', 'PROVINCIA_ID',
        'USER_HOMEBANKING_ID','DEBITOAUTOMATICO_ID','TIENE_MAIL_ID', 'SEXO_ID','CONTACTO_VALIDO_ID']]
y = df[['MORA_TARGET']]
y
```

```python
selector = SelectKBest(k=7)
selector.fit(x,y)
mask = selector.get_support()
x_cols = x.columns[mask]
x_cols
```

## Creacion de modelo de clasificacion

```python
datosEntrenamiento = df[x_cols]
etiquetaObjetivo = y
```

```python
x_train, x_test, y_train, y_test = train_test_split(datosEntrenamiento,etiquetaObjetivo, test_size=0.3)
print("Los datos para entrenamiento son :",x_train.shape)
print("Los datos para prueba son :",x_test.shape)
print("Los eti para entrenamiento son :",y_train.shape)
print("Los eti para prueba son :",y_test.shape)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
modelRF = RandomForestClassifier()
modelRF.fit(x_train, y_train)
y_predRF = modelRF.predict(x_test)
matrizRF = confusion_matrix(y_test,y_predRF)
print(classification_report(y_test, y_predRF))
```

```python
from sklearn.naive_bayes import GaussianNB
modelNB = GaussianNB()
modelNB.fit(x_train, y_train)
y_pred = modelNB.predict(x_test)
matrizG = confusion_matrix(y_test,y_pred)
print(classification_report(y_test, y_pred))
```
